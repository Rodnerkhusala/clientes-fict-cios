```python
# 1. EXTRAÇÃO
import pandas as pd

# Lendo o arquivo original
df = pd.read_csv("clientes_ficticios.csv", encoding="latin-1")
print("Dados originais:")
print(df)

```

```python
# 2. TRANSFORMAÇÃO
# Normalizar nomes (tudo maiúsculo)
df["Nome"] = df["Nome"].str.upper()

# Remover espaços dos números de cartão
df["Cartão"] = df["Cartão"].str.replace(" ", "")

# Criar coluna 'Agência' fictícia
df["Agência"] = ["001", "002", "003", "004"]

# Criar coluna 'Saldo Inicial' fictício
df["Saldo_Inicial"] = [5000, 3000, 7000, 4500]

print("\nDados transformados:")
print(df)

```

```python
# 3. CARREGAMENTO
# Salvar em novo CSV
df.to_csv("clientes_transformados.csv", index=False, encoding="latin-1")

print("\nArquivo 'clientes_transformados.csv' criado com sucesso!")

```
